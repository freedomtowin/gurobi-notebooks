
<h1 id="Milk-Collection-Problem">Milk Collection Problem<a class="anchor-link" href="https://gurobi.github.io/modeling-examples/milk_collection/milk_collection.html#Milk-Collection-Problem">¶</a></h1><h2 id="Objective-and-Prerequisites">Objective and Prerequisites<a class="anchor-link" href="https://gurobi.github.io/modeling-examples/milk_collection/milk_collection.html#Objective-and-Prerequisites">¶</a></h2><p>The Milk Collection Problem is an example of a capacitated vehicle routing problem. In this problem, one tanker lorry with limited capacity, collects milk from a set of farms, and the goal is to identify the optimal route for the tanker. This problem is formulated as a binary optimization problem using the Gurobi Python API and solved with the Gurobi Optimizer.</p>
<p>This model is example 23 from the fifth edition of Model Building in Mathematical Programming by H. Paul Williams on pages 278-281 and 336-337.</p>
<p>This modeling example is at the advanced level, where we assume that you know Python and the Gurobi Python API and you have  advanced knowledge of building mathematical optimization models. Typically, the objective function and/or constraints of these examples are complex or require advanced features of the Gurobi Python API.</p>

<h2 id="Problem-Description">Problem Description<a class="anchor-link" href="https://gurobi.github.io/modeling-examples/milk_collection/milk_collection.html#Problem-Description">¶</a></h2><p>A small milk processing company is committed to collecting milk from 20 farms and taking it back to the depot for processing. The company has one tanker lorry with the capacity to carry 80 000 liters of milk. Eleven of the farms are small and need a collection only every other day. The other nine farms need a collection every day. The positions of the farms in relation to the depot (numbered 1) are given in the following table together with their collection requirements.</p>
<p><img alt="farmCoordinates" src="./milk_collection_files/farmCoordinates.PNG"/></p>
<p>The goal is to find the optimal route for the tanker lorry on each day, bearing in mind that it has to:</p>
<ol>
<li>Visit all the ‘every day’ farms. </li>
<li>Visit some of the ‘every other day’ farms. </li>
<li>Work within its capacity. </li>
</ol>
<p>On alternate days, it must again visit the ‘every day’ farms and also visit the ‘every other day’ farms not visited on the
previous day.</p>

<h2 id="Python-Implementation">Python Implementation<a class="anchor-link" href="https://gurobi.github.io/modeling-examples/milk_collection/milk_collection.html#Python-Implementation">¶</a></h2><p>We import the Gurobi Python Module and other Python libraries.</p>


```

import gurobipy as gp
from gurobipy import GRB

import math
from itertools import combinations,permutations,product

# tested with Python 3.7.0 & Gurobi 9.1.0

```


<h3 id="Input-data">Input data<a class="anchor-link" href="https://gurobi.github.io/modeling-examples/milk_collection/milk_collection.html#Input-data">¶</a></h3><p>We define all the input data for the model.</p>


```

# Create a dictionary to capture the farm coordinates (10 miles) and collection requirements (1,000).

farms, coordinates, collect  = gp.multidict({
    0: [(0,0),0],
    1: [(-3,3),5],
    2: [(1,11),4],
    3: [(4,7),3],
    4: [(-5,9),6],
    5: [(-5,-2),7],
    6: [(-4,-7),3],
    7: [(6,0),4],
    8: [(3,-6),6],
    9: [(-1,-3),5],
    10: [(0,-6),4],
    11: [(6,4),7], 
    12: [(2,5),3],
    13: [(-2,8),4],
    14: [(6,10),5],
    15: [(1,8),6],
    16: [(-3,1),8],
    17: [(-6,5),5],
    18: [(2,9),7],
    19: [(-6,-5),6],
    20: [(5,-4),6]
})

# List of farm  including the depot
farms = [*range(0,21)]

# List of farms that are visited everyday
everyDay = [*range(0,10)]

# List of farms that are visited each other day
otherDay = [*range(10,21)]

# List of day types
dayType = [1,2]

# Tanker lorry capacity (1,000)
tankerCap = 80

# Every day farms requirements
everyDayReq = 0
for i in everyDay:
    everyDayReq += collect[i]

```


<h3 id="Preprocessing">Preprocessing<a class="anchor-link" href="https://gurobi.github.io/modeling-examples/milk_collection/milk_collection.html#Preprocessing">¶</a></h3><p>The following function calculates the distance of each pair of farms combination. We also define the data structures to define the decision variables, constraints, and objective function.</p>


```

# Compute pairwise distance matrix
# numpy linalg norm = euclidean n=2

def distance(city1, city2):
    c1 = coordinates[city1]
    c2 = coordinates[city2]
    diff = (c1[0]-c2[0], c1[1]-c2[1])
    return math.sqrt(diff[0]*diff[0]+diff[1]*diff[1])

dist = {(c1, c2): distance(c1, c2) for c1, c2 in product(farms, farms) if c1 != c2}

# Generate the edges for each day type
list_ijk = []
for i,j in dist.keys():
    for k in dayType:
        tp = i,j,k
        list_ijk.append(tp)
ijk = gp.tuplelist(list_ijk)

# List of other day farms and day type
list_ik = []
for i in otherDay:
    for k in dayType:
        tp = i,k
        list_ik.append(tp)
ik = gp.tuplelist(list_ik)

```


<h3 id="Callback-Definition">Callback Definition<a class="anchor-link" href="https://gurobi.github.io/modeling-examples/milk_collection/milk_collection.html#Callback-Definition">¶</a></h3><p>The following function determines lazy constraints to eliminate sub-tours.</p>


```

# Callback - use lazy constraints to eliminate sub-tours

def subtourelim(model, where):
    if where == GRB.Callback.MIPSOL:
        # make a list of edges selected in the solution
        vals = model.cbGetSolution(model._vars)
        # Selecting the farms visited during day 1
        selected1 = gp.tuplelist((i, j) for i, j, kk in model._vars.keys() 
                                 if (kk == 1 and vals[i, j, kk] > 0.5) )
        # find a cycle in the selected edge list of day 1
        nodes1 = set()
        for i,j in selected1:
            nodes1.add(i)
            nodes1.add(j)
        tour1 = subtour(selected1)
        if len(tour1) < len(nodes1):
            # add subtour elimination constr. for farms visited during day 1 and day 2
            model.cbLazy(gp.quicksum(model._vars[i, j, 1] for i, j in combinations(tour1, 2))
                         <= len(tour1)-1)
            model.cbLazy(gp.quicksum(model._vars[i, j, 2] for i, j in combinations(tour1, 2))
                         <= len(tour1)-1)
        # Selecting the farms visited during day 2
        selected2 = gp.tuplelist((i, j) for i, j, kk in model._vars.keys() 
                                 if (kk == 2 and vals[i, j, kk] > 0.5) )
        # find a cycle in the selected edge list of day 2
        nodes2 = set()
        for i,j in selected2:
            nodes2.add(i)
            nodes2.add(j)
        tour2 = subtour(selected2)
        if len(tour2) < len(nodes2):
            # add subtour elimination constr. for farms visited during day 1 and day 2
            model.cbLazy(gp.quicksum(model._vars[i, j, 2] for i, j in combinations(tour2, 2))
                         <= len(tour2)-1)
            model.cbLazy(gp.quicksum(model._vars[i, j, 1] for i, j in combinations(tour2, 2))
                         <= len(tour2)-1)

```


<h3 id="Finding-A-Cycle">Finding A Cycle<a class="anchor-link" href="https://gurobi.github.io/modeling-examples/milk_collection/milk_collection.html#Finding-A-Cycle">¶</a></h3><p>The following function determines a cycle not connected to the depot.</p>


```

# Given a tuplelist of edges, find the shortest subtour

def subtour(edges):
    nodes = set()
    for i,j in edges:
        nodes.add(i)
        nodes.add(j)
    unvisited = list(nodes)
    cycle = list(nodes) 
    while unvisited:  
        thiscycle = []
        neighbors = unvisited
        while neighbors:
            current = neighbors[0]
            thiscycle.append(current)
            unvisited.remove(current)
            neighbors = [j for i, j in edges.select(current, '*')
                         if j in unvisited]
        if len(thiscycle) <= len(cycle):
            cycle = thiscycle # New shortest subtour
    return cycle

```


<h3 id="Model-Deployment">Model Deployment<a class="anchor-link" href="https://gurobi.github.io/modeling-examples/milk_collection/milk_collection.html#Model-Deployment">¶</a></h3><p>We now determine the capacitated vehicle routing model for the milk collection problem by defining decision variables, constraints, and objective function.</p>


```

# Create the model object m
m = gp.Model('MilkCollection.lp')

# Decision ariables: 

# Edge variables = 1, if farm 'i' is adjacent to farm 'j' on the tour of day type 'k'.
vars = m.addVars(ijk, vtype=GRB.BINARY, name='x')


# Other day variables = 1, if farm 'i' is visited on the tour of day type 'k'.
other_var = m.addVars(ik, vtype=GRB.BINARY, name='y') 

# Symmetry constraints.
for i,j in dist.keys(): 
    for k in dayType:
        m.addConstr(vars[j, i, k] == vars[i, j, k], name='symmetry')  # edge in opposite direction

# Every day constraints: two edges incident to an every day farm on tour of day type 'k'. 
m.addConstrs((gp.quicksum(vars[tp] for tp in ijk.select(i,'*',k) ) 
              == 2 for i in everyDay for k in dayType  ), name='everyDay')

# Other day constraints: two edges incident to an other day farm on tour of day type 'k'.
m.addConstrs((gp.quicksum(vars[tp] for tp in ijk.select(i,'*',k) ) 
              == 2*other_var[i,k] for i in otherDay for k in dayType ), name='otherDay')

# Tanker capacity constraint.
m.addConstrs(( gp.quicksum(collect[i]*other_var[i,k] for i in otherDay ) <= tankerCap-everyDayReq for k in dayType ),
             name='tankerCap')

# Other day farms are visited on day type 1 or 2.
otherDayFarms = m.addConstrs((other_var[i,1] + other_var[i,2] == 1 for i in otherDay), name='visited')

# Avoid symmetric alternative solutions
other_var[11,1].lb = 1

# Objective function: minimize total distance travel

m.setObjective((gp.quicksum(0.5*dist[i,j]*vars[i,j,k] for i,j,k in ijk)), GRB.MINIMIZE)

```

```

# Optimize the model

m._vars = vars
m.Params.lazyConstraints = 1
m.optimize(subtourelim)

```
